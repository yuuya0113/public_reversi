```python
import tkinter as tk
import random



#クリックイベント

def click(event):
    global A,B,is_start,is_run
    xx = -1
    if not is_run:
        return
    else:
        is_start = True
        button["text"]="リセット"
        x,y = -1,-1;
        if B == p_color:
            xy = cp()
            xx = xy[0]
            yy = xy[1]
        else:
            if event.widget["text"] != "●":
                (xx,yy) = xy_read(event)
        if xx < 0:
                console["text"] = ("そこに置くことはできません。")
        elif check(xx,yy,True):
            new_text(f"{A}が{str_number(xx,yy)}に置きました。")
            board_write()
            change_turn()
            if check_pass():
                change_turn()
                if check_pass():
                    game_end()
                    return
                else:
                    new_text(f"[{B}はパスです。]")
            if is_run:
                if A == p_color:
                    console["text"] = f"{A}/あなたの版です。石を置く場所をクリックしてください。"
                else:
                    console["text"] = (f"{A}/相手の版です。盤面クリックで進みます。")
        else:
            console["text"] = ("そこに置くことはできません。")

#ボタンイベント
def push(event):
    global is_start,is_run,board,A,B,p_color
    if not is_start:
        p_color = "白";
        button["text"] = "リセット"
        is_start = True
        console["text"] =  "黒/相手の番です。盤面をクリックしてください。"
        return
    else:
        new_text("リセットしました。");
        board = [[0 for _ in range(8)] for _ in range(8*(rank+1))]
        board[3][3] = "黒"
        board[4][4] = "黒"
        board[3][4] = "白"
        board[4][3] = "白"
        A = "黒"
        B = "白"
        p_color = "黒"
        is_run = True
        for x in range(8):
            for y in range(8):
                number = str(x) + str(y)
                cell[x][y].config(foreground = "#00a000", text = number)
        board_write()
        button["text"] = "白で開始"
        console["text"] = "黒/あなたの番です。石を置く場所をクリックしてください。"
        is_start = False
        return
    
#エンターイベント
def enter(event):
    if event.widget["bg"] == "#00a000":
        event.widget["bg"] = "#00b000"
    if event.widget["foreground"] == "#00a000":
        event.widget["foreground"] = "#00b000"

#リーヴイベント
def leave(event):
    if event.widget["bg"] == "#00b000":
        event.widget["bg"] = "#00a000" 
    if event.widget["foreground"] == "#00b000":
        event.widget["foreground"] = "#00a000" 

#クリック位置取得
def xy_read(event):
    xx = int(int(event.widget["text"])//10)
    yy = int(int(event.widget["text"])%10)
    return (xx,yy)

#設置場所取得
def str_number(x,y):
    num = ["1","2","3","4","5","6","7","8"];
    cha = ["a","b","c","d","e","f","g","h"];
    return cha[y]+num[x]

#ログ生成
def new_text(text):
    log3["text"] = log2["text"]
    log2["text"] = log1["text"]
    log1["text"] = text
    return

#ターン交代
def change_turn():
    global A,B
    A,B = B,A
    return

#盤面描写
def board_write():
    for x in range(8):
        for y in range(8):
            if board[x][y] == "黒":
                cell[x][y].config(text="●", foreground="black")
            elif board[x][y] == "白":
                cell[x][y].config(text="●", foreground="white")
    return

#配置可否チェック
def check(i,j,rev=False):
    global A,B
    flag = False
    if cell[i][j]["text"] == "●":
        return False
    if (i < 7) and (0 < j) and (board[i+1][j-1] == B):
        if reverse(i+1,j-1,1,rev):
            flag = True
            if rev:
                board[i+1][j-1] = A;
                
    if (i < 7) and (board[i+1][j] == B):
        if reverse(i+1,j,2,rev):
            flag = True
            if rev:
                board[i+1][j] = A;

    if (i < 7) and (j < 7) and (board[i+1][j+1] == B):
        if reverse(i+1,j+1,3,rev):
            flag = True 
            if rev:
                board[i+1][j+1] = A;
                
    if (0 < j) and (board[i][j-1] == B):
        if reverse(i,j-1,4,rev):
            flag = True
            if rev:
                board[i][j-1] = A;

    if (j < 7) and (board[i][j+1] == B):
        if reverse(i,j+1,6,rev):
            flag = True
            if rev:
                board[i][j+1] = A;

    if (0 < i) and (0 < j) and (board[i-1][j-1] == B):
        if reverse(i-1,j-1,7,rev):
            flag = True
            if rev:
                board[i-1][j-1] = A;

    if (0 < i) and (board[i-1][j] == B):
        if reverse(i-1,j,8,rev):
            flag = True;
            if rev:
                board[i-1][j] = A;
                
    if (0 < i) and (j < 7) and (board[i-1][j+1] == B):
        if reverse(i-1,j+1,9,rev):
            flag = True;
            if rev:
                board[i-1][j+1] = A;

    if flag and rev:
        board[i][j] = A
    return flag

#リバース
def reverse(i,j,vec,rev):
    global A
    global B
    
    if vec == 1:
        i,j = i+1,j-1
    elif vec == 2:
        i,j = i+1,j
    elif vec == 3:
        i,j = i+1,j+1
    elif vec == 4:
        j = j-1
    elif vec == 6:
        j = j+1
    elif vec == 7:
        i,j = i-1,j-1
    elif vec == 8:
        i,j = i-1,j
    elif vec == 9:
        i,j = i-1,j+1
    
    if (0 <= i <= 7) and (0 <= j <= 7):
        if board[i][j] == B:
            if reverse(i,j,vec,rev):
                if rev:
                    board[i][j] = A
                return True;
        if board[i][j] == A:
            return True
    return False;

#パスチェック
def check_pass():
    for xz in range(8):
        for yz in range(8):
            if check(xz,yz,False):
                return False;
    return True;




#cpアルゴリズム
def cp(ccount=0):
    max_score,rx,ry,even_count = 99,-1,-1,1
    t_board = [[0 for _ in range(8)] for _ in range(8*(rank+1))]
    
    for x in range(8):
        for y in range(8):
            t_board[x][y] = board[x][y]
      
    for x in range(8):
        for y in range(8):
            if check(x,y,True):
                change_turn()
                can_put = 0
                for xz in range(8):
                    for yz in range(8):
                        if(xz,yz,False):
                            can_put += 1
                if can_put < max_score:
                    rx = x;
                    ry = y;
                    max_score = can_put;
                    even_count = 1;
                elif can_put == max_score:
                    even_count += 1
                    if random.randint(1,even_count) == 1:
                        rx = x;
                        ry = y;
                        max_score = can_put;
                        
                change_turn();
            for xz in range(8):
                for yz in range(8):
                     board[xz][yz]=t_board[xz][yz]        
    return (rx,ry,max_score)

#ゲーム終了処理
def game_end():
    global board,is_run
    w = 0
    b = 0
    for x in range(8):
        for y in range(8):
            if board[x][y] == "黒":
                b += 1
            if board[x][y] == "白":
                w += 1
                
    new_text(f"[黒：{b}][白：{w}]")
    if (p_color =="黒" and b > w) or (p_color == "白" and w > b):
        console["text"] = "あなたの勝ちです。"
    elif b == w:
        console["text"] = "引き分けです。";
    else:
        console["text"] = "あなたの負けです。";
    is_run = False;


# ウィンドウ生成・初期設定

window = tk.Tk()
window.config(bg="white")
window.title("リバーシ")
frame1,frame2 = tk.Frame(),tk.Frame()
frame1.pack()
frame2.pack(fill="x")
is_run = True
is_start = False

cell = [[],[],[],[],[],[],[],[]]
for x in range(8):
    for y in range(8):
        number = str(x) + str(y)
        cell[x].append(tk.Label(frame1,
                                anchor="center",
                                relief = "ridge", 
                                bg = "#00a000",
                                width = 2, 
                                height = 1,
                                foreground = "#00a000",
                                text = number,
                                font=("", "25", "")))
        cell[x][y].grid(row = x, column = y)
cell[3][3].config(foreground = "black", text="●")
cell[4][4].config(foreground = "black", text="●")
cell[3][4].config(foreground = "white", text="●")
cell[4][3].config(foreground = "white", text="●")


button = tk.Button(frame2, text = "白で開始", anchor="center", width=7)
button.pack(side="left")
console = tk.Label(frame2, text="黒/あなたの版です。石を置く場所をクリックしてください。", anchor="w", justify='left',bg="lightgray")
console.pack(side="top", fill="x")
log1 = tk.Label(frame2, anchor="w", justify='left', foreground="#202020")
log1.pack(side="top", fill="x")
log2 = tk.Label(frame2, anchor="w", justify='left', foreground="#404040")
log2.pack(side="top", fill="x")
log3 = tk.Label(frame2, anchor="w", justify='left', foreground="#606060")
log3.pack(side="top", fill="x")

rank = 0
board = [[0 for _ in range(8)] for _ in range(8*(rank+1))]
debug_count = 0
A = "黒"
B = "白"
p_color = "黒"
j,k = -1,-1
board[3][3],board[3][4],board[4][3],board[4][4]="黒","白","白","黒"

#イベント
for x in range(8):
    for y in range(8):
        cell[x][y].bind("<1>",click)
        cell[x][y].bind("<Enter>", enter)
        cell[x][y].bind("<Leave>", leave)
button.bind("<1>", push)
window.mainloop() 
```
